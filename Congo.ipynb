```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait       
from selenium.webdriver.common.by import By  
from selenium.webdriver.common.action_chains import ActionChains     
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from msedge.selenium_tools import EdgeOptions
from selenium.webdriver import Chrome, ChromeOptions
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import csv
import time
import re
import json
import pandas as pd
import numpy as np
import os
from selenium.webdriver.common.keys import Keys

option1=Options()

prefs = {"profile.default_content_setting_values.notifications" : 2}
option1.add_experimental_option("prefs",prefs)
option1.add_argument('--disable-popup-blocking')
option1.add_argument('--disable-notifications')
option1.add_argument("--no-sandbox")
driver = Chrome(service=Service(ChromeDriverManager().install()), options=option1)
# service = Service(executable_path='â€ªC:/Users/Yu Bai/OneDrive/msedgedriver.exe',options=option1)
# options = EdgeOptions()
# options.use_chromium = True
# driver = webdriver.Edge(service=service, options=options)
#driver = Edge(options)
#driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()
driver.implicitly_wait(10)

#Notification and Maximize window
    

for i in range(1,117):
        url = 'https://www.carmunu.com/en/buy-cars?page='+str(i)
        driver.get(url)
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, "html.parser")
             
        for link in soup.findAll("div", class_=re.compile("ad-specification", re.I)):
                try:
                    Location=link.find('div', class_='location').text
                
                except Exception as e:
                    Location = None
                try:
                    Name=link.find('div', class_='specification-section').text
                   
                except Exception as e:
                    Name = None
                try:      
                    Price=link.find('div', attrs={'class':'ad-vehicle-price'}).text
                  
                except Exception as e:
                    Price = None
                try:
                    Capacity=link.find('div', attrs={'class':'engine-capacity'}).text
                   
                except Exception as e:
                    Capacity = None
                try:
                    transmission=link.find('div', attrs={'class':'transmission'}).text
            
                except Exception as e:
                    transmission = None
                try:
                    mileage=link.find('div', attrs={'class':'ad-vehicle-mileage'}).text
                   
                except Exception as e:
                    mileage = None
                try:
                    Desc = link.find(name='div', attrs={'class': "md-hidden sort-ad-description"}).text
                except:
                    Desc=None
                try:
                    url_path = link.div.a.get('href')
                    urlpath = "https://www.carmunu.com"+url_path
                except Exception as e:
                    urlpath=None
                     
                if (Name==None) and (Location == None) and (Price == None) and (Capacity==None) and (mileage == None) and (urlpath==None) and  (transmission==None) and (Desc==None):
                    continue
                else:
                    with open('carmunucongo.csv','a', encoding='utf-8', newline='') as file:
                        writer=csv.writer(file)
                        file_is_empty = os.stat('carmunucongo.csv').st_size == 0
                        if file_is_empty:
                              writer.writerow(['Name','Location','price',"Capacity",'mileage',"urlpath",'transmission','Desc'])

                        writer.writerow([Name,Location,Price,Capacity,mileage,urlpath,transmission,Desc])
        file.close()

```
