```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait       
from selenium.webdriver.common.by import By  
from selenium.webdriver.common.action_chains import ActionChains     
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv
import time
import re
import json
import pandas as pd
import numpy as np
import os
from selenium.webdriver.common.keys import Keys

option1=Options()

prefs = {"profile.default_content_setting_values.notifications" : 2}
option1.add_experimental_option("prefs",prefs)
option1.add_argument('--disable-popup-blocking')
option1.add_argument('--disable-notifications')

#option1.add_argument("--no-sandbox")

service = Service(executable_path='C:/Users/Nandini Dulam/OneDrive/WebScraping/chromedriver.exe',chrome_options=option1)
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()
driver.implicitly_wait(10)

#Notification and Maximize window
    

for i in range(1,334):
    
        url = 'https://www.truecar.com/new-cars-for-sale/listings/location-russia-oh/?page='+str(i)
        
        driver.get(url)
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, "html.parser")
             
        for link in soup.findAll("div", class_=re.compile("flex w-full flex-col", re.I)):
                    try:
                        name=link.find('div', class_='vehicle-card-top').text
                        #print(name)
                    except Exception as e:
                        name = None
                    try:
                        Name=link.find('div', class_='vehicle-card-top').text
                       # print(Name)
                    except Exception as e:
                        Name = None
                    try:      
                        Price=link.find('div', attrs={'class':'pr-3 vehicle-card-bottom-pricing-primary'}).text
                       # print(Price)
                    except Exception as e:
                        Price = None
                    try:
                        Year=link.find('span', attrs={'class':'vehicle-card-year text-xs'}).text
                        #print(Year)
                    except Exception as e:
                        Year = None
                    try:
                        loc=link.find('div', attrs={'class':'vehicle-card-location mt-1 text-xs'}).text
                       # print(loc)
                    except Exception as e:
                        loc = None
                    try:
                        url_path = link.a.get('href')
                        #print(url_path)
                        urlpath = "https://www.truecar.com"+url_path
                        #print(urlpath)
                        #Transmission,Desc = get_url(urlpath)
                    except Exception as e:
                        urlpath=None
                    driver.get(urlpath)
                    soup_url1 = BeautifulSoup(driver.page_source, "html.parser")
                    for soup_url in soup_url1.findAll("div", class_=re.compile("mt-4 mb-2 lg:mt-0", re.I)):
                           # print(soup_url)
                            try:        
                                Information = soup_url.find(name='div', attrs={'class': 'row pt-3'}).text 
                                #print(Information)
                            except:
                                Information=None
#                             try:        
#                                 Transmission = soup_url.find(name='i', attrs={'class': "flaticon-gearshift transmission"}).text 
#                                 #print(Transmission)
#                             except:
#                                 Transmission=None
#                             try:
#                                 Desc = soup_url.find(name='div', attrs={'class': "mt-4 mb-2 lg:mt-0"}).text
#                                 #print(Desc)
#                             except:
#                                  Desc=None
                    if (Name==None) and (name == None) and (Price == None) and (Year == None) and (loc == None) and (urlpath==None) and (Information==None):
                        continue
                    else:
                        with open('truecarrussia.csv','a', encoding='utf-8', newline='') as file:
                            writer=csv.writer(file)
                            file_is_empty = os.stat('truecarrussia.csv').st_size == 0
                            if file_is_empty:
                                  writer.writerow(['Name',"Year","loc",'name','price',"urlpath",'Information'])

                            writer.writerow([Name,Year,loc,name,Price,urlpath,Information])
        file.close()

```
