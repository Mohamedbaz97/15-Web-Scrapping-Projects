```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait       
from selenium.webdriver.common.by import By  
from selenium.webdriver.common.action_chains import ActionChains     
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv
import time
import re
import json
import pandas as pd
import numpy as np
import os
from selenium.webdriver.common.keys import Keys

option1=Options()

prefs = {"profile.default_content_setting_values.notifications" : 2}
option1.add_experimental_option("prefs",prefs)
option1.add_argument('--disable-popup-blocking')
option1.add_argument('--disable-notifications')

#option1.add_argument("--no-sandbox")

service = Service(executable_path='C:/Users/Nandini Dulam/OneDrive/WebScraping/chromedriver.exe',chrome_options=option1)
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()
driver.implicitly_wait(10)

#Notification and Maximize window
    

for i in range(1,38):
    
        url = 'https://autochek.africa/ng/cars-for-sale?page_number='+str(i)
        
        driver.get(url)
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, "html.parser")
             
        for link in soup.findAll("div", class_=re.compile("MuiBox-root css-qym38b", re.I)):
                    try:
                        name=link.find('h6', class_='MuiTypography-root MuiTypography-h6 css-1g399u0').text
                        #print(name)
                    except Exception as e:
                        name = None
                    try:
                        Name=link.find('h6', class_='MuiTypography-root MuiTypography-h6 css-1g399u0').text
                        #print(Name)
                    except Exception as e:
                        Name = None
                    try:      
                        Price=link.find('p', attrs={'class':'MuiTypography-root MuiTypography-body1 css-1bztvjj'}).text
                        #print(Price)
                    except Exception as e:
                        Price = None
#                     try:
#                         Cate=link.find('div', attrs={'class':'new-offer-category'}).text
#                         #print(Cate)
#                     except Exception as e:
#                         Cate = None
#                     try:
#                         loc=link.find('div', attrs={'class':'new-card-place'}).text
#                         #print(loc)
#                     except Exception as e:
#                         loc = None
                    try:
                        url_path = link.div.a.get('href')
                        urlpath = "https://autochek.africa/"+url_path
                        #print(urlpath)
                        #Transmission,Desc = get_url(urlpath)
                    except Exception as e:
                        urlpath=None
                    driver.get(urlpath)
                    soup_url1 = BeautifulSoup(driver.page_source, "html.parser")
                    for soup_url in soup_url1.findAll("div", class_=re.compile("MuiPaper-root MuiPaper-elevation MuiPaper-rounded MuiPaper-elevation0 css-rrc68p", re.I)):
                            try:        
                                Information = soup_url.find(name='div', attrs={'class': "MuiStack-root css-j7qwjs"}).text 
                               # print(Information)
                            except:
                                Information=None
#                             try:        
#                                 Transmission = soup_url.find(name='i', attrs={'class': "flaticon-gearshift transmission"}).text 
#                                 #print(Transmission)
#                             except:
#                                 Transmission=None
                            try:
                                Desc = soup_url.find(name='ul', attrs={'class': "MuiList-root MuiList-padding css-1ontqvh"}).text
                                #print(Desc)
                            except:
                                 Desc=None
                    if (Name==None) and (name == None) and (Price == None) and (url_path==None) and (Desc==None) and (Information==None):
                        continue
                    else:
                        with open('Autonigeria.csv','a', encoding='utf-8', newline='') as file:
                            writer=csv.writer(file)
                            file_is_empty = os.stat('Autonigeria.csv').st_size == 0
                            if file_is_empty:
                                  writer.writerow(['Name','name','price',"url_path",'Desc','Information'])

                            writer.writerow([Name,name,Price,url_path,Desc,Information])
        file.close()

```
