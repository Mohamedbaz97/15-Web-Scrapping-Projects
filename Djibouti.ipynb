```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait       
from selenium.webdriver.common.by import By  
from selenium.webdriver.common.action_chains import ActionChains     
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv
import time
import re
import json
import pandas as pd
import numpy as np
import os
from selenium.webdriver.common.keys import Keys

option1=Options()

prefs = {"profile.default_content_setting_values.notifications" : 2}
option1.add_experimental_option("prefs",prefs)
option1.add_argument('--disable-popup-blocking')
option1.add_argument('--disable-notifications')
option1.add_argument("--no-sandbox")
service = Service(executable_path='C:/Users/Nandini Dulam/OneDrive/WebScraping/chromedriver.exe',chrome_options=option1)
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()
driver.implicitly_wait(10)

#Notification and Maximize window
    

for i in range(1,39):
        url = 'https://www.cardjibouti.com/en/buy-cars-in-Ali%20Sabieh%20Region-Import%20-%20Dubai?page='+str(i)
        driver.get(url)
        time.sleep(2)
        soup = BeautifulSoup(driver.page_source, "html.parser")
             
        for link in soup.findAll("div", class_=re.compile("ad-specification", re.I)):
                try:
                    Location=link.find('div', class_='location').text
                    print(Location)
                except Exception as e:
                    Location = None
                try:
                    Name=link.find('div', class_='specification-section').text
                    print(Name)
                except Exception as e:
                    Name = None
                try:      
                    Price=link.find('div', attrs={'class':'ad-vehicle-price'}).text
                    print(Price)
                except Exception as e:
                    Price = None
                try:
                    Capacity=link.find('div', attrs={'class':'engine-capacity'}).text
                    print(Capacity)
                except Exception as e:
                    Capacity = None
                try:
                    transmission=link.find('div', attrs={'class':'transmission'}).text
                    print(transmission)
                except Exception as e:
                    transmission = None
                try:
                    mileage=link.find('div', attrs={'class':'ad-vehicle-mileage'}).text
                    print(mileage)
                except Exception as e:
                    mileage = None
                try:
                    Desc = link.find(name='div', attrs={'class': "md-hidden sort-ad-description"}).text
                    print(Desc)
                except:
                    Desc=None
                try:
                    url_path = link.div.a.get('href')
                    urlpath = "https://www.cardjibouti.com/"+url_path
                    print(urlpath)
                   
                except Exception as e:
                    urlpath=None
                     
                if (Name==None) and (Location == None) and (Price == None) and (Capacity==None) and (mileage == None) and (urlpath==None) and  (transmission==None) and (Desc==None):
                    continue
                else:
                    with open('cardjibouti.csv','a', encoding='utf-8', newline='') as file:
                        writer=csv.writer(file)
                        file_is_empty = os.stat('cardjibouti.csv').st_size == 0
                        if file_is_empty:
                              writer.writerow(['Name','Location','price',"Capacity",'mileage',"urlpath",'transmission','Desc'])

                        writer.writerow([Name,Location,Price,Capacity,mileage,urlpath,transmission,Desc])
        file.close()

```
